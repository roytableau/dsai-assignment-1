# Feature Engineering

From the cleaned data, we derived:
*   Seniority from position level
*   Salary bands from min/max salary
*   Demand metrics based on vacancy counts
*   Popular job titles grouped into common roles

# Analytics Consideration

Because categories were exploded, some measures could be duplicated.

For example, a single professional-level job might belong to both Information Technology and Customer Service.

Therefore, we used averages instead of sums when analysing salaries to avoid overstating values.
